```python
import math
import numpy as np
import matplotlib.pyplot as plt

data_set = [(-3, 9),
    (-2, 4),
    (-1, 1),
    (0, 0),
    (1, 1),
    (2, 4),
    (3, 9)]
n = len(data_set)

# Starting weights
w_0 = 0
w_1 = 0
w_2 = 0

learning_rate = 0.005
epochs = 200

# Setup the plot before the loop
plt.figure(figsize=(10, 6))

# Extract data points for plotting
x_points = np.array([pt[0] for pt in data_set])
y_points = np.array([pt[1] for pt in data_set])
x_line = np.linspace(min(x_points), max(x_points), 100)

for i in range(epochs):
    if i in [0, 5, 10, 20, 40, 80, 100, 120, 140, 160, 180, 200]:
        print(f'w0: {w_0}')
        print(f'w1: {w_1}')
        print(f'w2: {w_2}')
        y_line = w_0 + w_1 * x_line + w_2 * x_line**2

        alpha_val = 0.15 + (i / epochs) * 0.5
        plt.plot(x_line, y_line, color='blue', alpha=alpha_val, linestyle='--',
                 label=f'Epoch {i}')


    loss = 0
    dw_0 = 0
    dw_1 = 0
    dw_2 = 0

    for x, y in data_set:
        prediction = w_0 + w_1 * x + w_2 * x**2
        error = y - prediction
        loss += error**2
        dw_0 += -2 * error
        dw_1 += -2 * error * x
        dw_2 += -2 * error * x**2

    loss /= n
    dw_0 /= n
    dw_1 /= n
    dw_2 /= n

    w_0 = w_0 - learning_rate * dw_0
    w_1 = w_1 - learning_rate * dw_1
    w_2 = w_2 - learning_rate * dw_2

y_final_line = w_0 + w_1 * x_line + w_2 * x_line**2
plt.plot(x_line, y_final_line, color='darkblue', linewidth=2.5,
         label=f'Final Model (Epoch {epochs})')


plt.scatter(x_points, y_points, color='red', s=80, label='Data Points', zorder=5)


plt.title('Visualizing how the Regression Line Changes Over Time')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()


plt.show()

print(f'w0: {w_0}')
print(f'w1: {w_1}')
print(f'w2: {w_2}')
```
